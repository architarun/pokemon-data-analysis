### Data Pre-processing

```python
total = poke.isnull().sum().sort_values(ascending=False)
percent = (poke.isnull().sum()/poke.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent],axis=1, keys=['Total','Percent'])
missing_data.head(41)
```

```python
poke['type2'].fillna('None', inplace=True)
```
